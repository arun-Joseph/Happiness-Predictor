```python
import numpy as np
import pandas as pd
import csv
```

```python
with open('2015.csv', 'r') as csvfile:
    spamreader = csv.reader(csvfile, delimiter=',')
    data_2015 = list(spamreader)
    data_2015 = np.asarray(data_2015)
    df_2015 = pd.DataFrame(data=data_2015[1:], columns=data_2015[0])
    df_2015.drop(columns=['Happiness Rank', 'Region', 'Standard Error'], axis=1, inplace=True)
    df_2015.columns = ['Country', 'Score', 'Economy', 'Family', 'Health', 'Freedom', 'Trust', 'Generosity', 'Dystopia']
    df_2015.set_index('Country', inplace=True)
    df_2015.sort_index(inplace=True)
    df_2015 = df_2015.astype(float)
```

```python
df_2015
```

```python
with open('2016.csv', 'r') as csvfile:
    spamreader = csv.reader(csvfile, delimiter=',')
    data_2016 = list(spamreader)
    data_2016 = np.asarray(data_2016)
    df_2016 = pd.DataFrame(data=data_2016[1:], columns=data_2016[0])
    df_2016.drop(columns=['Happiness Rank', 'Region', 'Lower Confidence Interval', 'Upper Confidence Interval'],
                 axis=1, inplace=True)
    df_2016.columns = ['Country', 'Score', 'Economy', 'Family', 'Health', 'Freedom', 'Trust', 'Generosity', 'Dystopia']
    df_2016.set_index('Country', inplace=True)
    df_2016.sort_index(inplace=True)
    df_2016 = df_2016.astype(float)
```

```python
with open('2017.csv', 'r') as csvfile:
    spamreader = csv.reader(csvfile, delimiter=',')
    data_2017 = list(spamreader)
    data_2017 = np.asarray(data_2017)
    df_2017 = pd.DataFrame(data=data_2017[1:], columns=data_2017[0])
    df_2017.drop(columns=['Happiness.Rank', 'Whisker.high', 'Whisker.low'], axis=1, inplace=True)
    df_2017.columns = ['Country', 'Score', 'Economy', 'Family', 'Health', 'Freedom', 'Generosity', 'Trust', 
                       'Dystopia']
    df_2017 = df_2017.reindex(columns=['Country', 'Score', 'Economy', 'Family', 'Health', 'Freedom', 'Trust', 
                                       'Generosity', 'Dystopia'])
    df_2017.set_index('Country', inplace=True)
    df_2017.sort_index(inplace=True)
    df_2017 = df_2017.astype(float)
```
