```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_pickle('happiness_data.pkl')
df
```

```python
df.info()
```

```python
df.describe()
```

```python
df.corr()
```

```python
sns.kdeplot(df['Score'])
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Score Distribution Analysis')
plt.show()
```

```python
sns.kdeplot(df['Family'])
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Score(Family\) Distribution Analysis')
plt.show()
```

```python
for x in range(1,34):
    sns.kdeplot(df['Ind '+ str(x)])
    plt.xlabel('Values')
    plt.ylabel('Frequency')
    plt.title('Ind '+ str(x) + ' Distribution Analysis')
    plt.show()
```

```python
for x in range(1,34):
    plt.ylabel('Score')
    plt.xlabel('Ind '+ str(x))
    plt.title('Score vs Ind '+ str(x))
    plt.scatter(df['Ind '+ str(x)],df['Score'])
    plt.show()
```

```python
for x in range(1,34):
    for y in range(1,34-x):
        plt.ylabel('Ind '+ str(y))
        plt.xlabel('Ind '+ str(x))
        plt.title('Ind '+ str(y) +' vs Ind '+ str(x))
        plt.scatter(df['Ind '+ str(x)],df['Ind '+ str(y)])
        plt.show()
```
